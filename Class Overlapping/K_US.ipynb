```python
import pandas as pd
import statistics
```

```python
dataset = pd.read_csv("cm1.csv")
dataset
```

```python
# label encoding 0 and 1 for target variable
from sklearn.preprocessing import LabelEncoder as LE
L_Defects = LE()

New_data = dataset
New_data["defects"] = L_Defects.fit_transform(New_data["defects"])
New_data.head()
```

```python
# creating x and y variables that will store NonDefective and Defective
X=New_data.loc[New_data['defects']==0]
Y=New_data.loc[New_data['defects']==1]
X
```

```python
# checking which is  NonDefective and Defective based on their frequency of occurence
if len(X)>len(Y):
    NonDefective=X
    Defective=Y
else :
    Defective=X
    NonDefective=Y
NonDefective
```

```python
#data preprocessing
X= New_data.drop('defects', axis='columns')
Y= New_data['defects']
```

```python
To_Be_Removed = pd.DataFrame(columns=['index','membership_count'])    
To_Be_Removed

NonDefective.index[0]
```

```python
import numpy as np

To_Be_Removed = pd.DataFrame(columns=['index','membership_count'])
data = []
distances_NonDefective = []
NonDefective_neighbors_ids = []

for i in range(0,len(Defective)):
    # KNN search function to find n nearest points using eucliedian dist and
    # placing them in colonial neigbhors
    new_data_point = np.array(X.loc[Defective.index[i]])
    distances = np.linalg.norm(X - new_data_point, axis=1)

    # K=10 because IR value is 10 for the given dataset cm1
    k = 10
    K_neighbors = distances.argsort()[1:k+1]

    print(K_neighbors)
    # finding the NonDefective colonial neighbors from all the colonial neighbors that
    # were found earlier
    for i in K_neighbors:
        if(Y[i]==NonDefective.loc[NonDefective.index[0]]['defects']):
            data.append(X.loc[i])
            distances_NonDefective.append(distances[i])
            NonDefective_neighbors_ids.append(i)

Neighbors = pd.DataFrame()
Neighbors[0] = NonDefective_neighbors_ids
Neighbors[1] = distances_NonDefective
print(Neighbors)
print(Neighbors.loc[0][0])
```

```python
#calculating the membership count
for j in range(0,len(Neighbors)):
    i = Neighbors.loc[j][0]
    if To_Be_Removed['index'].isin([i]).any().any():
        To_Be_Removed.loc[To_Be_Removed['index']==i,'membership_count']+=1
    else:
        new_row = {'index':i,'membership_count':1}
        To_Be_Removed=To_Be_Removed.append(new_row, ignore_index=True)
To_Be_Removed
```

```python
# checking those that have membership count more than 1
To_Be_Removed = To_Be_Removed.loc[To_Be_Removed['membership_count']>1]
To_Be_Removed
```

```python
#sorting the final elimination based on value of index because while droping in 
#the last step indexes will be changed , so to avoid wrong instance elimination 
# we sort them in decreasing order of their indices
To_Be_Removed = To_Be_Removed.sort_values(["index"], ascending=False)
To_Be_Removed
```

```python
#dropping those instances that were chosen for final elimination
for i in To_Be_Removed['index']:
    dataset.drop(dataset.index[int(i)],inplace=True)
    
dataset
```

```python
# Saving our new improved dataset into a new .csv file
dataset.to_csv("D.csv")
```
