```python
import pandas as pd
import statistics
import numpy as np
```

```python
dataset = pd.read_csv("cm1.csv")
dataset
```

```python
from sklearn.preprocessing import LabelEncoder as LE
L_Defects = LE()

New_data = dataset
New_data["defects"] = L_Defects.fit_transform(New_data["defects"])
New_data.head()
```

```python
X=New_data.loc[New_data['defects']==0]
Y=New_data.loc[New_data['defects']==1]
```

```python
if len(X)>len(Y):
    majority=X
    minority=Y
else :
    minority=X
    majority=Y
```

```python
X= New_data.drop('defects', axis='columns')
Y= New_data['defects']
```

```python
majority.head()
```

```python
import numpy as np
data = []
distances_majority = []
R = []


for i in range(0,len(minority)):
    new_data_point = np.array(X.loc[minority.index[i]])
    distances = np.linalg.norm(X - new_data_point, axis=1)

    k = 3
    n_neighbor_ids = distances.argsort()[1:k+1]
    
    

    for j in n_neighbor_ids:
        if(Y[i]==majority.loc[majority.index[0]]['defects']):
            data.append(X.loc[i])
            distances_majority.append(distances[i])
            R.append(i)


R= list(set(R))
R.sort(reverse=True)
for i in R:
    dataset.drop(dataset.index[int(i)],inplace=True)
```
