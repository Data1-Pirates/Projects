```python
import pandas as pd
import statistics
```

```python
dataset = pd.read_csv("cm1.csv")
dataset
```

```python
# label encoding 0 and 1 for target variable
from sklearn.preprocessing import LabelEncoder as LE
L_Defects = LE()

New_data = dataset
New_data["defects"] = L_Defects.fit_transform(New_data["defects"])
New_data.head()
```

```python
# creating x and y variables that will store clean and buggy
X=New_data.loc[New_data['defects']==0]
Y=New_data.loc[New_data['defects']==1]
```

```python
# checking which is clean and buggy based on their frequency of occurence
if len(X)>len(Y):
    clean=X
    buggy=Y
else :
    buggy=X
    clean=Y
```

```python
#data preprocessing
X= New_data.drop('defects', axis='columns')
Y= New_data['defects']
```

```python

to_eliminate = pd.DataFrame(columns=['index','membership_count'])    
to_eliminate
```

```python
import numpy as np

to_eliminate = pd.DataFrame(columns=['index','membership_count'])


for i in range(0,len(buggy)):
    # KNN search function to find n nearest points using eucliedian dist and
    # placing them in colonial neigbhors
    new_data_point = np.array(X.loc[buggy.index[i]])
    distances = np.linalg.norm(X - new_data_point, axis=1)

    # K=10 because IR value is 10 for the given dataset cm1
    k = 10
    colonial_neighbor_ids = distances.argsort()[1:k+1]


    data = []
    distances_clean = []
    clean_neighbors_ids = []
    # finding the clean colonial neighbors from all the colonial neighbors that
    # were found earlier
    for i in colonial_neighbor_ids:
        if(Y[i]==clean.loc[clean.index[0]]['defects']):
            data.append(X.loc[i])
            distances_clean.append(distances[i])
            clean_neighbors_ids.append(i)
          
  
    # Calculating median of all the distances of clean neighbors
    med = statistics.median(distances_clean)

    df = pd.DataFrame()
    df[0] = clean_neighbors_ids
    df[1] = distances_clean
    
    print(df)

    Nominated_for_elimination = []
    #if the distance is less than median , then nominate those for elimination
    for i in range(0,len(clean_neighbors_ids)):
        if(df.loc[i][1]<=med):
            Nominated_for_elimination.append(df.loc[i][0])
    print(Nominated_for_elimination)
    #calculating the membership count
    for i in Nominated_for_elimination:
        if to_eliminate['index'].isin([i]).any().any():
            to_eliminate.loc[to_eliminate['index']==i,'membership_count']+=1
        else:
            new_row = {'index':i,'membership_count':1}
            to_eliminate=to_eliminate.append(new_row, ignore_index=True)
```

```python
# checking those that have membership count more than 1
to_eliminate.loc[to_eliminate['membership_count']>1]
```

```python
dataset.head()
```

```python
# Storing those that have more than 1 membership count in final_elimination

final_elimination = to_eliminate.loc[to_eliminate['membership_count']>1]

final_elimination
```

```python
#sorting the final elimination based on value of index because while droping in 
#the last step indexes will be changed , so to avoid wrong instance elimination 
# we sort them in decreasing order of their indices
sorted_df = final_elimination.sort_values(["index"], ascending=False)
sorted_df
```

```python
#dropping those instances that were chosen for final elimination
for i in sorted_df['index']:
    dataset.drop(dataset.index[int(i)],inplace=True)
    
dataset
```

```python
# Saving our new improved dataset into a new .csv file
dataset.to_csv("New-Data.csv")
```
